```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')
import hvplot.pandas
```

```python
if 'data' not in pn.state.cache.keys():
    df= pd.read_csv('https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv')
    pn.state.cache['data'] = df.copy()
else:
    df = pn.state.cache['data']
```

Data Preprocessing

```python
df=df.fillna(0)
df['gdp_per_capita'] = np.where(df['population']!=0, df['gdp'] / df['population'] , 0)
```

```python
df['gdp_per_capita']
```

```python
idf = df.interactive()
```

CO2 Emission by Continent over time

```python
year_slider = pn.widgets.IntSlider(name = 'Year Slider', start = 1750, end = 2020, step = 5, value = 1851)
year_slider
```

```python
yaxis_CO2 = pn.widgets.RadioButtonGroup(name = 'Y-axis', options = ['co2', 'co2 per capita'], button_type = 'success') 
```

```python
continents = ['Asia', 'North America', 'South America', 'Europe', 'Australia', 'World']

co2_pipeline = ( idf[(idf.year<=year_slider) & (idf.country.isin(continents))].groupby(['country', 'year'])[yaxis_CO2].mean().to_frame().reset_index().sort_values(by = 'year').reset_index(drop = True))
```

```python
co2_pipeline
```

```python
co2_plot = co2_pipeline.hvplot(x = 'year', by = 'country', min_width = 100, y=yaxis_CO2,line_width=2, title="CO2 emission by continent", width = 800)

co2_plot
```

```python
co2_table = co2_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 20, sizing_mode='stretch_width') 
co2_table
```

```python
co2_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.year == year_slider) &
        (~ (idf.country.isin(continents)))
    ]
    .groupby(['country', 'year', 'gdp_per_capita'])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)


co2_vs_gdp_scatterplot_pipeline
```

```python
co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(x='gdp_per_capita', 
                                                                y='co2', 
                                                                by='country', 
                                                                size=80, kind="scatter", 
                                                                alpha=0.7,
                                                                legend=False, 
                                                                height=500, 
                                                                width=500,
                                                                title='Relation Between CO2 and GDP per Capita')
co2_vs_gdp_scatterplot
```

```python
yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['coal_co2', 'oil_co2', 'gas_co2'], 
    button_type='success'
)

continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_source_bar_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.country.isin(continents_excl_world))
    ]
    .groupby(['year', 'country'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(kind='bar', 
                                                     x='country', 
                                                     y=yaxis_co2_source, 
                                                     title='CO2 source by continent')
co2_source_bar_plot
```

```python
template = pn.template.FastListTemplate(
    title='World CO2 emission Dashboard', 
    sidebar=[pn.pane.Markdown("# CO2 Emissions and Climate Change"), 
             pn.pane.Markdown("####  emissions are the primary driver of global climate change. It’s widely recognised that to avoid the worst impacts of climate change, the world needs to urgently reduce emissions. But, how this responsibility is shared between regions, countries, and individuals has been an endless point of contention in international discussions."), 
             pn.pane.PNG('climate_day.png', sizing_mode='scale_both'),
             pn.pane.Markdown("## Settings"),   
             year_slider],
    main=[pn.Row(pn.Column(yaxis_CO2, 
                           co2_plot.panel(width=900), margin=(0,25)), 
                 co2_table.panel(width=500)), 
          pn.Row(pn.Column(co2_vs_gdp_scatterplot.panel(width=600), margin=(0,25)), 
                 pn.Column(yaxis_co2_source, co2_source_bar_plot.panel(width=600)))],
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)
# template.show()
template.servable();
```
